```python
import pymysql

database = pymysql.connect(host='localhost',
                           user='root',
                           password='Gangurde@156',
                           database='demo')

Cursor = database.cursor()
sql = "insert into student (prn,name,subject) values (%s,%s,%s)"
val = [("101","Pushkar","DBMS"),("102","Aarohi","SQL")]
Cursor.executemany(sql,val)

database.commit()

print(Cursor.rowcount,"Inserted")
```

```python
import pymysql

database = pymysql.connect(host='localhost',
                           user='root',
                           password='Gangurde@156',
                           database='demo')

Cursor = database.cursor()
sql = "delete from student where prn=%s and name=%s and subject=%s"
val = ("101","Pushkar","DBMS")
Cursor.execute(sql,val)

database.commit()

print(Cursor.rowcount,"deleted")
Cursor.close()
database.close()
```
